```python
import Bio.PDB.PDBParser
import sys

# Use QUIET=True to avoid lots of warnings...
parser = Bio.PDB.PDBParser(QUIET=True)
structure = parser.get_structure("1KCW", "1kcw.pdb")
model = structure[0]
achain = model['A']

cysresidues = []

#Voy por cada residuo hasta encontrar CYS
for residue in achain:
    if residue.get_resname() == 'CYS':
        cysresidues.append(residue)
```

```python
# Se crean 3 listas vacias(esto con el fin de no repetir resultados) para guardar los indices de c1 y c2 y también guardar su distancia.
list_distance=[]
list_c1index=[]
list_c2index=[]
for c1 in cysresidues:
    c1index = c1.get_id()[1]
    for c2 in cysresidues:
        c2index = c2.get_id()[1]
#El primer "and" se usa para evitar cys con el mismo indice y valor,
# el 2° y 3° "and" se utilizan para que los datos que esten en una lista no se reflejen en la otra. Así tenemos 2 listas unicas,con valores unicos.
        if ((c1['SG'] - c2['SG']) < 8.0) and ((c1index!=c2index)) and ((c2index in list_c1index)==False) and ((c1index in list_c2index)==False) :
            list_c1index.append(c1index)
            list_c2index.append(c2index)
            list_distance.append(round(c1['SG'] - c2['SG'],2))

#Se imprimen los datos obtenidos.
for i in range (len(list_c1index)):
    print("possible di-sulfide bond:", end=" ")
    print("Cys",list_c1index[i],"-", end=" ")
    print("Cys",list_c2index[i], end=" ")
    print(list_distance[i])
```

```python
import Bio.PDB.PDBParser
import sys

# Use QUIET=True to avoid lots of warnings...
parser = Bio.PDB.PDBParser(QUIET=True)
structure = parser.get_structure("1HPV","1ce0.pdb")

achain = structure[0]['A']
bchain = structure[0]['B']
```

```python
from Bio.PDB.Polypeptide import is_aa
```

```python
# Busco en el carbono de la "colita"
for res1 in achain:
    # Se revisa que sea un aa y además que sea especificamente Lisina.
    if is_aa(res1) and (res1.get_resname()=="LYS"):
        r1ca = res1['NZ']
        r1ind = res1.get_id()[1]
        r1sym = res1.get_resname()
        # Busco el carbono de la "colita" de la otra cadena
        for res2 in bchain:
            # Se revisa que sea un aa y además que sea especificamente Lisina.
            if is_aa(res2) and (res2.get_resname()=="LYS") :
                r2ca = res2['NZ']
                r2ind = res2.get_id()[1]
                r2sym = res2.get_resname()
                
                #las distancias entre lisinas deben ser menores a 12 A
                if (r1ca - r2ca) < 12:
                    print("Residues",r1sym,r1ind,"in chain A", end=" ")
                    print("and",r2sym,r2ind,"in chain B", end=" ")
                    print("are close to each other:",round(r1ca-r2ca,2))
                    
```

<h2 id="problem">Ejercicio 4</h2>
<p>Automaticen para buscar los siguientes genes en la base de datos nucleotide y guarden las secuencias de las proteínas en un archivo fasta: 
    
    
    Organismo: Notoryctes typhlops             Genes: PNOC y RUNX
    Organismo: Cypripedium guttatum            Genes: matK
    Organismo: Hepatitis B virus               Genes: X                     Aislado: migrant 467
    
    
</p>
</div>

```python
from Bio import Entrez
Entrez.email = "otokani@lcg.unam.mx"
#Para obtener todos los genes de todos los organismos se usan simplememte operaciones logicas con "AND" y "OR"
search_handle = Entrez.esearch(db="nucleotide",term="(Notoryctes typhlops AND (RUNX2 OR PNOC)) OR (Cypripedium guttatum AND matK) OR (Hepatitis B virus AND (X Aislado: migrant 467))  ", usehistory="y")
search_results = Entrez.read(search_handle)
search_handle.close()
```

```python
count =int(search_results["Count"])
```

```python
webenv = search_results["WebEnv"]
query_key = search_results["QueryKey"]
```

```python
from Bio import Entrez
import time
from urllib.error import HTTPError 

# Indicamos el numero de genes que vamos a recuperar
batch_size = 5

#archivo de salida
out_handle = open("genes.fasta", "w")

for start in range(0, count, batch_size):
    end = min(count, start+batch_size)
    print("Going to download record %i to %i" % (start+1, end))
    attempt = 1
    while attempt <= 3:
        try:
            fetch_handle = Entrez.efetch(db="nucleotide", rettype="fasta", retmode="text",
                                         retstart=start, retmax=batch_size,
                                         webenv=webenv, query_key=query_key)
            
            break
        except HTTPError as err:
            if 500 <= err.code <= 599:
                print("Received error from server %s" % err)
                print("Attempt %i of 3" % attempt)
                attempt += 1
                time.sleep(15)
            else:
                raise
                
    data = fetch_handle.read()
    fetch_handle.close()
    out_handle.write(data)
out_handle.close()
```
