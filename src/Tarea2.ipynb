<h2 id="problem">Ejercicio 1</h2>
<p>Un ORF inicia con un codón inicial y termina, ya sea con un codón final o al final de la cádena.
</p>
<p>Utilizando como input una cadena de ADN, obtener la cadena proteica de mayor longitud que puede ser traducida de cualquiera de sus ORF.
</p>
<p>Si más de una cadena protéica de longitud maxima existe, entonces podrás emitir cualquier solución.</p>
<h3 id="sample-dataset">Conjunto de datos de muestra</h3>
<div class="codehilite"><pre>AGCCATGTAGCTAACTCAGGTTACATGGGGATGACCCCGCGACTTGGATTAGAGTCTCTTTTGGAATAAGCCTGAATGATCCGAGTAGCATCTCAG
</pre></div>


<h3 id="sample-output">Resultado emitido de muestra</h3>
<div class="codehilite"><pre>MLLGSFRLIPKETLIQVAGSSPCNLS
</pre></div>
</div>

````python
from Bio.Seq import Seq
ADN_forward= Seq("AGCCATGTAGCTAACTCAGGTTACATGGGGATGACCCCGCGACTTGGATTAGAGTCTCTTTTGGAATAAGCCTGAATGATCCGAGTAGCATCTCAG")
ADN_reverse=ADN_forward.reverse_complement()
#Se usan dos listas. Una para guardar las proteinas y otra para guardar sus respectivos tamaños.
list_protein=[]
list_pro_length=[]

for ORF in range(3):
#Se hace un corte en la secuecnia forward y reverse dependiendo del marco de lectura correspondiente.
    strand_forward=ADN_forward[ORF:len(ADN_forward)]
    strand_reverse=ADN_reverse[ORF:len(ADN_forward)]
#Al hacer la traducción nos arroja la proteína con codones de paro y de inicio juntos en una misma secuencia.
    strand_forward=strand_forward.translate()
    strand_reverse=strand_reverse.translate()
#Dividimos la proteína (en proteínas más chicas) cada que encuentre un codon de paro.
    for protein in strand_forward.split("*"):
        #Volvemos a hacer cortes en las proteínas (en los codones de inicio) para tener una proteína "real".
        if protein.find("M")!=-1:
            start=protein[protein.find("M"):len(protein)]
            list_protein.append(start)
            list_pro_length.append(len(start))
            
            
    for protein in strand_reverse.split("*"):
        if protein.find("M")!=-1:
            start=protein[protein.find("M"):len(protein)]
            list_protein.append(start)
            list_pro_length.append(len(start))
#Cada proteína "real" se guarda en una lista al igual que su respectivo tamaño en la otra lista.Se obtiene la más grande y se imprime
index_max=(list_pro_length.index(max(list_pro_length)))
large_protein=list_protein[index_max]

print("Larger protein: {}".format(large_protein))
            
            
````

<div class="alert alert-block alert-info">
<h2 id="problem">Ejercicio 2</h2>
    
1. Teniendo como input un archivo fasta, crea un diccionario de todas las secuencias en tu atchivo FASTA.

2. Encuentra todos los codones en tu primer marco de lectura. 

3. Imprime los codones de cada secuncia en formato FASTA. 

   ```
   >my_seq_marco_1
   ATG TTC ATC
   ```


4. Imprime los codones de cada secuencia en los 6 marcos lectura.
</div>


````python
#seq.nt.fa
from Bio.Seq import Seq
from Bio import SeqIO
import re
print("Enter the path of your fasta file")
path=input()
#Se crea un diccionario con el archivo fasta
dictionary = SeqIO.to_dict(SeqIO.parse(path, 'fasta'))

for sequence in dictionary:
    #Se ingresa a cada secuencia del archivo fasta y se almacena esa secuencia en ADN_forward
    ADN_forward= dictionary[sequence].seq
    ADN_reverse=ADN_forward.reverse_complement()
    for ORF in range(3):
        #Dos listas donde se guardarán los codones de la secuencia reverse y forward.De cada marco de lectura
        codons_f=[]
        codons_r=[]
        strand_forward=ADN_forward[ORF:len(ADN_forward)]
        strand_reverse=ADN_reverse[ORF:len(ADN_forward)]
        print("\n>my_seq_marco_{}".format(ORF))
        #Se divide la secuencia forward en codones y posteriormente se imprimen.
        for codon in re.findall(r"(.{3})",str(strand_forward)):
            codons_f.append(codon)
        for codon in codons_f:
            print(codon,end="\t") 
        #Se divide la secuencia reverse en codones y posteriormente se imprimen.
        print("\n>my_seq_marco_-{}".format(ORF))
        for codon in re.findall(r"(.{3})",str(strand_reverse)):
            codons_r.append(codon)
        for codon in codons_r:
            print(codon,end="\t")
        
````

````python
from Bio import SeqIO
#Se lee el archivo de GenBank y se guarda la info. en una variable
gb_record= SeqIO.read("hepatitis.gb", "genbank")
#features[2]: ahí se encuentra "CDS". En CDS se encuentra la proteína que codifica la region de ADN.
#Accedemos con la llave "protein" y accesamos a la casilla 0 para que nos regrese la proteína.
protein=gb_record.features[2].qualifiers['translation'][0]
protein=protein.format("fasta")
print(protein)
#Creamos un archivo con la proteína obtenida en formato fasta.
with open("protein.fasta","w+") as file:
        file.write(">seq1\n")
        file.write(protein)
````

+ Calcula el centro geométrico (centroid) del aminoacido mencionado anteriormente e imprímelo (esto asume que las masas de los átomos son iguales)

Para calular el centro geométrico se necesitan sacar las coordenadas de todos los átomos del residuo y sacar el promedio.
</ul>

Con el ejemplo del residuo utilizado en **clase** debe darles:
```python
>>> centroid(residue)

````python
from Bio import PDB
parser = PDB.PDBParser(QUIET=True)
#Se parsea el archivo PDB y se consigue el structure. Se accede especificamente al residuo solicitado.
structure = parser.get_structure("protein", "4hhb.pdb")
residue=structure[0]["C"][20]
print("3D coordinates={}".format(residue["CA"].get_coord()))
#Para obtener el centroid se usa un "for" para sumar todos las coordenadas y posteriormente se divide entre el total de coordenadas.
n=0
atoms=0
for atom in residue:
                atoms = atom.get_coord()+atoms
                n+=1
centroid=atoms/n

print("centroid={}".format(centroid))
````
